### Seasonal decomposition: Identify seasonal patterns using decomposition techniques (e.g., using seasonal_decompose from statsmodels).

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#creating the dataframe to datetime index
climatedf.drop(columns=["Year","Month"],axis=1,inplace=True)
climatedf.head()
```

```python
climatedf.to_csv("data/climateml.csv")
```

```python
# We analyze averege temperature)
ts = climatedf['Average_Temperature_C']

# Perform decomposition on the monthly data, use period=12 (12 months in a year)
result = seasonal_decompose(ts, model='additive', period=12)
```

```python
# Or plot components individually 
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
fig.suptitle("Seasonal patterns of monthly Average Temperature")
result.observed.plot(ax=ax1, title='Observed')
result.trend.plot(ax=ax2, title='Trend')
result.seasonal.plot(ax=ax3, title='Seasonal')
result.resid.plot(ax=ax4, title='Residual')

plt.tight_layout()
plt.show()
```

```python
# We analyze total rainfall
ts2 = climatedf['Total_Rainfall_mm']

# Perform decomposition on the monthly data, use period=12 (12 months in a year)
result2 = seasonal_decompose(ts2, model='additive', period=12)
```

```python
# Plotting components individually 
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
fig.suptitle("Seasonal patterns of monthly Total Rainfall")
result2.observed.plot(ax=ax1, title='Observed')
result2.trend.plot(ax=ax2, title='Trend')
result2.seasonal.plot(ax=ax3, title='Seasonal')
result2.resid.plot(ax=ax4, title='Residual')

plt.tight_layout()
plt.show()
```

```python
# Extract the seasonal component
seasonal_component = result.seasonal

# Get the average seasonal pattern by month
monthly_seasonal = seasonal_component.groupby(seasonal_component.index.month).mean()

# Plot the seasonal pattern
plt.figure(figsize=(10, 4))
monthly_seasonal.plot(kind='bar')
plt.title('Average Seasonal Pattern for Temperature by Month')
plt.xlabel('Month')
plt.ylabel('Seasonal Component')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(alpha=0.3)
plt.show()
```

```python
# Extract the seasonal component
seasonal_component2 = result2.seasonal

# Get the average seasonal pattern by month
monthly_seasonal = seasonal_component2.groupby(seasonal_component.index.month).mean()

# Plot the seasonal pattern
plt.figure(figsize=(10, 4))
monthly_seasonal.plot(kind='bar')
plt.title('Average Seasonal Pattern for Rainfall by Month')
plt.xlabel('Month')
plt.ylabel('Seasonal Component')
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(alpha=0.3)
plt.show()
```

### 4. Machine Learning Model Development:
##### Objective: Predict future climate conditions based on historical data.

Tasks:

- Split data into training and testing sets.

- Select appropriate ML model(s) (e.g., Linear Regression, Random Forest) for prediction.

- Train the model on historical data.

- Evaluate model performance using metrics like RMSE (Root Mean Square Error), MAE (Mean Absolute Error).

- Output: Trained model capable of predicting future climate conditions in Tanzania.

```python
climateml = pd.read_csv("data/climateml.csv")
#climateml= climatedf.set_index("date")
climateml = climateml.sort_index()
climateml.head()
```

```python
climateml['date']= pd.to_datetime(climateml['date'])
```

```python
# ML Model: random forest regressor
climateml['year'] = climateml['date'].dt.year
climateml['month']= climateml['date'].dt.month
```

```python
climateml.head()
```

```python
climateml.info()
```

```python
# Cyclical encoding for months (helps model understand seasonality)
climateml['month_sin'] = np.sin(2 * np.pi * climateml['month'] / 12)
climateml['month_cos'] = np.cos(2 * np.pi * climateml['month'] / 12)
```

```python
#lag features for time series forecasting 
climateml['lag_1'] = climateml['Average_Temperature_C'].shift(1) # previous month
climateml['lag_12'] = climateml['Average_Temperature_C'].shift(12) #same month last year
```

```python
climateml.head()
```

```python
climateml= climateml.dropna()
                                              
```

```python
climateml = climateml.drop(columns=['date'], axis=1)
```

```python
climateml.head()
```

```python
len(climateml)
```

```python
#train-test split (80% train, 20%test)
train_size = int(0.8 * len(climateml))
train = climateml.iloc[:train_size]
test = climateml.iloc[train_size:]

# Separate features (X) and target (y)
X_train = train.drop(['Average_Temperature_C'], axis=1)  # Predictor variables
y_train = train['Average_Temperature_C']  # Target variable

X_test = test.drop(['Average_Temperature_C'], axis=1)
y_test = test['Average_Temperature_C']
```

```python
#ML model training using Random Forest regressor
from sklearn.ensemble import RandomForestRegressor

#initialize the model
rfmodel = RandomForestRegressor(
    n_estimators = 200,
    max_depth = 10,
    random_state = 42,
    n_jobs=1)
    
```

```python
# train the model
rfmodel.fit(X_train,y_train)
```

```python
# hyperparameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10]
}

gridsearch = GridSearchCV(rfmodel, param_grid, cv=6, scoring="neg_mean_squared_error")
gridsearch.fit(X_train, y_train)

best_rfchoice = gridsearch.best_estimator_
```

```python
print(best_rfchoice)
```

```python
#model evaluation - Prediction
y_pred = rfmodel.predict(X_test)
```

```python
#model evaluation - evaluation metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"MAE: {mae:.2f}")
print(f"R2: {r2:.2f}")
print(f"RMSE: {rmse:.2f}")
```

### Generate future dates and features


```python
# Find the last year/month in training data
last_year = climateml['year'].max()
last_month = climateml[climateml['year'] == last_year]['month'].max()

# Generate next 12 months
future_data = []
for month_ahead in range(1, 13):
    year = last_year + (last_month + month_ahead - 1) // 12
    month = (last_month + month_ahead - 1) % 12 + 1
    future_data.append({'year': year, 'month': month})

future_df = pd.DataFrame(future_data)
```

```python
#Add Cyclical Month Encoding

future_df['month_sin'] = np.sin(2 * np.pi * future_df['month'] / 12)
future_df['month_cos'] = np.cos(2 * np.pi * future_df['month'] / 12)
```

```python
future_df.head()
```

```python
#incorporate lag features
# Get last known temperature
last_temp = climateml[(climateml['year'] == last_year) & (climateml['month'] == last_month)]['Average_Temperature_C'].values[0]

# Initialize with lag-1
future_df['lag_1'] = last_temp


for idx, row in future_df.iterrows():
    same_month_last_year = climateml[(climateml['year'] == row['year'] - 1) & (climateml['month'] == row['month'])]
    future_df.at[idx, 'lag_12'] = same_month_last_year['Average_Temperature_C'].mean() if not same_month_last_year.empty else np.nan

# Fill any missing lag-12 values (e.g., if no history exists)
future_df['lag_12'] = future_df['lag_12'].fillna(climateml['Average_Temperature_C'].median())
```
